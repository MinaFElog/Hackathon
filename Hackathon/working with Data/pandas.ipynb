```python
sid = [12547, 12548, 12549]
name = ['John', 'Ben', 'Emily']
field = ['Computer Science', 'Law', 'Finance']
```

```python
# First Approach
students = pd.DataFrame({'SID': sid, 'Name': name, 'Field': field})
print(students)
```

```python
# Second Approach
students2 = pd.DataFrame([sid, name, field], columns=['First', 'Second', 'Third'])
print(students2)
```
