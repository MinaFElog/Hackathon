```python
import re
```

```python
sent = 'Alice was born in 1996'
sent = re.sub(r'\d', '', sent)
print(sent)
```

```python
sent = 'Alice was born in 1996, in England.'
sent = re.sub(r'[a-z]', '', sent)
print(sent)
```

```python
sent = 'Alice was born    in 1996.'
sent = re.sub(r'\s+', ' ', sent)
print(sent)
```

```python
sent = '22 years ago, he came to UK, when he was 28'
sent = re.sub(r'^[0-9]{1,}', '', sent)
print(sent)
```

```python
sent = '22 years ago, he came to UK, when he was 28'
sent = re.sub(r'[0-9]{1,}$', '', sent)
print(sent)
```

```python
sent = 'Alice was born in 1996, in England.'
sent = re.sub(r'[^a-zA-Z]', ' ', sent)
print(sent)
```

```python
sent = 'Alice was born in 1996'
print(re.match(r'.*', sent))
```

```python
date = '12.02.2022'
if (re.match(r'[0-9]{2}(.|/)[0-9]{2}(.|/)[0-9]{2,4}', date)):
    print('Matched.')
else:
    print('Not matched!')
```

```python
sent = 'Alice was born in 1996'
print(re.findall(r'[0-9]{1,}', sent))
```

```python
sent = 'Alice was born in 1996'
print(re.findall(r'alice', sent))
```

```python
sent = 'Alice was born in 1996'
print(re.findall(r'alice', sent, flags=re.I))
```
