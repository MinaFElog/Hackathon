```python
my_string = '5 4 6 3'
```

```python
my_list = my_string.split(' ')
print(my_list)
```

```python
my_list = list(map(int, my_list))
print(my_list)
```

```python
print(my_list[0])
```

```python
print(my_list[1:3])
```

```python
print(my_list[1:])
```

```python
print(my_list[-1])
```

```python
print(*my_list, sep='\n')
```

```python
my_list.append(6)
print(my_list)
```

```python
my_list.insert(0, 7)
print(my_list)
```

```python
new_list = [100, 200, 300]
my_list += new_list
print(my_list)
```

```python
my_list = [5, 4, 6, 3]
new_list = [100, 200, 300]
my_list.extend(new_list)
print(my_list)
```

```python
my_list = [1, 2, 4, 3, 5]
my_list.sort()
print(my_list)
```

```python
my_list = [1, 2, 4, 3, 5]
sorted(my_list)
print(my_list)
```

```python
my_list = [1, 2, 4, 3, 5]
my_list.sort(reverse = True)
print(my_list)
```

```python
my_list = [1, 2, 4, 3, 5]
print(my_list[::-1])
```

```python
my_list = [1, 2, 4, 3, 5]
my_list.reverse()
print(my_list)
```

```python
my_list = ['I', 'love', 'coding!']
my_string = ' '.join(my_list)
print(my_string)
```
