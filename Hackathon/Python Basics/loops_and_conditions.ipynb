```python
# for loop
for i in range(5):
    print(i)
```

```python
# for loop
for i in range(10):
    print(i)
    if i == 4:
        break
```

```python
# nested for loop
for i in range(3):
    for j in range(100, 102):
        print(i, '+', j,'=', i+j)
```

```python
# while loop
i = 0
while i < 5:
    print(i)
    i += 1
```

```python
# while loop
while True:
    print(1)
```

```python
# if statement
i = 1
j = 1
if i == j:
    print('Equal')
```

```python
# if statement
i = 1
j = 0
if i > j:
    print('Greater')
elif i <= j:
    print('Smaller')
```
